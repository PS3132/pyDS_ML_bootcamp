# Style and color

```python
import  matplotlib as plt
%matplotlib inline
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.countplot(x="sex", data=tips)
sns.set_style("darkgrid")
```

```python
sns.countplot(x="sex", data=tips)
sns.set_style("white")
sns.despine(left=False, bottom=True)
```

```python
sns.set_context(context="poster",font_scale=1)
sns.countplot(x="sex", data=tips)
```

```python
sns.countplot(x="sex", data=tips, palette="winter")
# for more color https://matplotlib.org/examples/color/colormaps_reference.html
```
