## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## import and normal fig

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
ax.set_title("Title")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

## Inside fig and axis

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(z, y, "r")
ax2.plot(y, z, "r")
```

## Inside fig and multiple axis with direction limitation

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x, x*2, "b", lw="1")
ax1.plot(x, z, "g")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])


ax2.plot(x, z, "b", lw="1")
ax2.set_title("Zoom")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_xlim([20,22])
ax2.set_ylim([300,500])
```

## subplot in table form

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,"b--", lw="3")
axes[1].plot(x,z,"r", lw="3")
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(10,2))
axes[0].plot(x,y,"b--", lw="3")
axes[1].plot(x,z,"r", lw="3")
```
